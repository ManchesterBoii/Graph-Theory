##  <span style="color:red">  Définition
Un  graphe  non orienté (resp. orienté) d'ordre $n \neq 0$, $G = (X,A)$, est donné par un ensemble $X=\lbrace s_1,s_2,\cdots,s_n\rbrace $ de $n$ éléments appelés sommets et  par un ensemble $A=\lbrace a_1,a_2,,\cdots,a_k \rbrace$ d'arêtes (resp. arcs).
