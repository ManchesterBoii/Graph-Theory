##  <span style="color:red">  Définition
Un  graphe  non orienté (resp. orienté) d'ordre $n \neq 0$, $G = (X,A)$, est donné par un ensemble $X=\lbrace s_1,s_2,\cdots,s_n\rbrace $ de $n$ éléments appelés sommets et  par un ensemble $A=\lbrace a_1,a_2,,\cdots,a_k \rbrace$ d'arêtes (resp. arcs).

`````python
G1.categorical_product?
'''Return the tensor product of "self" and "other".

   The tensor product of G and H is the graph L with vertex set V(L)
   equal to the Cartesian product of the vertices V(G) and V(H), and
   ((u,v), (w,x)) is an edge iff - (u, w) is an edge of self, and -
   (v, x) is an edge of other.'''
`````
