```python
# REPONSE :
def test(x, graph):
    E = []
    G_vert = graph.vertices()
    print(G_vert)
    M = (matrix.identity(len(G_vert)) + graph.adjacency_matrix() )**len(G_vert) 
    
    print(M)
 
    for y in G_vert:
        if(y!=x):
            if M[x,y] > 0:
                E.append(y)
    for y in E:
        if M[y,x] > 0:
            E.remove(y)
            
    return len(E) == 0, E
```

## Exercice 2 :<span style= "color:#00B8DE">  Connexité et forte connexité.

#### 1. Implémentez la fonction `` Est_connexe() `` qui retourne vrai si le graphe est connexe, faux sinon.

```python
# REPONSE :
def est_connexe(G):
    g = G.to_dictionary()
    
    visited = set()
    stack = [G.vertices()[0]]
    
    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            stack.extend(set(g[node])-visited)
            
    
    return len(visited)== len(G.vertices())
    
    
```

#### 4. Implémentez la fonction `` Est_strictement_connexe() `` qui retourne ``vrai`` si le graphe est fortement connexe, ``faux`` sinon.

```python
# REPONSE :
def Est_strictement_connexe(G):
    graph = G.to_dictionary()
    
    for i in G.vertices():
        visited = [False] * len(G.vertices())
 
        DFS(graph, i, visited)
 
        for b in visited:
            if not b:
                return False
 
    return True
    
    
    
def DFS(graph, v, visited):
 
    visited[v] = True
 
    for u in graph[v]:
        if not visited[u]:
            DFS(graph, u, visited)
 
 

    
```

```python
# REPONSE :
def Color_sommets_WP(G):
    g= G.to_dictionary()
    groups = []
    used = []
    deg = {i: len(g[i]) for i in g.keys() }
    orderDeg = list({k: v for k, v in reversed(sorted(deg.items(), key=lambda item: item[1]))}.keys())
    s = list({k: v for k, v in reversed(sorted(deg.items(), key=lambda item: item[1]))}.keys())
    
    for node in orderDeg:
        if node in used:
            continue
        group = [node]
        gv = []
        gv.extend(g[node])
        used.append(node)
        for u in used:
            if u not in s:
                continue
            s.remove(u)
            
        for node2 in s:
            if (node2 not in gv):
                group.append(node2)
                used.append(node2)
                gv.extend(g[node2])
        
                    
        groups.append(group)
        
        
    
    return groups
        
        
    
```
