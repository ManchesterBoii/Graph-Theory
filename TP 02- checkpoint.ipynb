#### 1. Représenter graphiquement le graphe orienté dont les sommets sont les entiers de 2 à 30, de sorte qu'il y ait une arête allant du sommet $n$ au sommet $m$ si et seulement si $n$ divise $m$. (On disposera les sommets de manière ciruclaire.)

````python
sommet=list(range(2,31))
G8=DiGraph()
G8.add_vertices(sommet)
aret = []

for m in sommet:
    for n in range(2,m):
        if m%n == 0:
            aret.append((n,m))
G8.add_edges(aret)
````

#### 2. Construire une fonction `ConstructGrapheDivise` qui prend en argument un nombre entier non nul $n$ et rend et affiche un graphe $G$ orienté dont les sommets sont les entiers de 2 à $n$, de sorte qu'il y ait une arête allant du sommet $i$ au sommet $j$ si et seulement si $i$ divise $j$ . (On disposera les sommets de manière ciruclaire.)

````python
def ConstructGrapheDivise(n):
    sommet=list(range(2,n+1))
    G=DiGraph()
    G.add_vertices(sommet)
    aret = []

    for j in sommet:
        for i in range(2,j):
            if j%i == 0:
                aret.append((i,j))
    G.add_edges(aret)
    return G.plot(figsize=15,vertex_size=300,layout='circular')
    
ConstructGrapheDivise(15).show() 
ConstructGrapheDivise(7).show()  
````
